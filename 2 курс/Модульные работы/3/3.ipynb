h) правило 3х сигм

```python
X = df_after.iloc[:, :-1].values

# Кластеризация методом ближнего соседа
single_clustering = AgglomerativeClustering(n_clusters=2, linkage='single')
single_clusters = single_clustering.fit_predict(X)
single_score = silhouette_score(X, single_clusters)

# Кластеризация методом дальнего соседа
complete_clustering = AgglomerativeClustering(n_clusters=2, linkage='complete')
complete_clusters = complete_clustering.fit_predict(X)
complete_score = silhouette_score(X, complete_clusters)

# Кластеризация методом центра тяжести
ward_clustering = AgglomerativeClustering(n_clusters=2, linkage='ward')
ward_clusters = ward_clustering.fit_predict(X)
ward_score = silhouette_score(X, ward_clusters)

# Кластеризация методом средней связи
average_clustering = AgglomerativeClustering(n_clusters=2, linkage='average')
average_clusters = average_clustering.fit_predict(X)
average_score = silhouette_score(X, average_clusters)

# Вывод индексов силуэта
print("Silhouette scores:")
print("Single linkage: ", single_score)
print("Complete linkage: ", complete_score)
print("Ward linkage: ", ward_score)
print("Average linkage: ", average_score)

# Визуализация кластеров
plt.scatter(X[:, 0], X[:, 1], c=single_clusters)
plt.title("Single linkage")
plt.show()

plt.scatter(X[:, 0], X[:, 1], c=complete_clusters)
plt.title("Complete linkage")
plt.show()

plt.scatter(X[:, 0], X[:, 1], c=ward_clusters)
plt.title("Ward linkage")
plt.show()

plt.scatter(X[:, 0], X[:, 1], c=average_clusters)
plt.title("Average linkage")
plt.show()
```
